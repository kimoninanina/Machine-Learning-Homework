```python
import pandas as pd
import numpy as np
import sklearn
data_list=[]
for line in open("C:\\Users\\muzzl\\Downloads\\lenses.data"):
    features=line.split()
    intFeatures=[]
    for num in features:
        newnum=int(num)
        intFeatures.append(newnum)
    data_list.append(intFeatures[1:])
df=pd.DataFrame(data_list,columns=['age',"prescription","astigmatic","tearRate","Class"])
df.tail()
```

```python
X = df[['age','prescription','astigmatic','tearRate']].values
y = df['Class']

print('feature X size', X.size)
print('class label y size', y.size)
print('Class labels:', np.unique(y))
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=1, stratify=y)

print('X-train size', X_train.size)
print('X-test size', X_test.size)
print('y-train size', y_train.size)
print('y-test size', y_test.size)
```

```python
print('Labels counts in y:', np.bincount(y))
print('Labels counts in y_train:', np.bincount(y_train))
print('Labels counts in y_test:', np.bincount(y_test))
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
sc.fit(X_train)
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier

tree = DecisionTreeClassifier(criterion='gini',                               
                                max_depth=4, 
                              random_state=1)
tree.fit(X_train, y_train)
```

```python
print('Accuracy Score on the test data: ',tree.score(X_test,y_test) )
```

```python
from pydotplus import graph_from_dot_data
from sklearn.tree import export_graphviz

dot_data = export_graphviz(tree,
                           filled=True, 
                           rounded=True,
                           class_names=['hard_contact','soft_contact','no_contact'],
                           feature_names=['age','prescription','astigmatic','tearRate'],
                           out_file=None) 
graph = graph_from_dot_data(dot_data) 
graph.write_png('tree.png') 
```
